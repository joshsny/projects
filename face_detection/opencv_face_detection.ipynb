```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import requests
from bs4 import BeautifulSoup
import re
%matplotlib inline
```

```python
img_raw = cv2.imread('mandrill_colour.png')
```

```python
print(type(img_raw))
img_raw.shape
```

```python
img_rgb = cv2.cvtColor(img_raw, cv2.COLOR_BGR2RGB) # Need to convert from cv2's BGR format to matplotlib's RGB format, try without this to see what it looks like

# draw a line
cv2.line(img_rgb, (0,0), (288,288), (255,0,0), 5) # Draw a red line of width 5 pixels across the image

#draw a rectangle
cv2.rectangle(img_rgb, (288,0), (200, 200), (0,0,255), 3) # Draw a blue rectangle with width 3 pixels

#draw a circle
cv2.circle(img_rgb, (200,200), 30, (0,255,0), -1) # Draw a filled in circle

# write some text
cv2.putText(img_rgb, 'Joshua S.', (150,288), 0, 0.5, (255, 255, 255), 2, cv2.LINE_AA) 

plt.imshow(img_rgb)
```

```python
page = requests.get('https://github.com/joshsny')
contents = page.content

print(contents)
```

```python
soup = BeautifulSoup(contents, 'html.parser')
raw_titles = soup.findAll(["h1", "h2","h3", "span"], {'class' : 'Name'})
titles = []
for title in raw_titles:
    titles.append(title.string)

print(titles)
```
